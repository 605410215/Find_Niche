```python
import pandas as pd
import json as js
import numpy as np
import math
from scipy.stats import entropy
import time
```

```python
df=pd.read_csv('finish_msno_entropy',sep=',')
df
```

```python
df2=df.groupby(['artist_id','msno'])['count_number'].sum()
df2=df2.reset_index()
df2=df2.sort_values('artist_id',ascending=False)
df2
```

```python
#df2.to_csv('artist_count_num.csv',sep=',',encoding='utf-8',index=False)
```

```python
df3=df2.groupby('msno')['count_number'].sum()
df3=df3.reset_index()
df3
```

```python
df3.rename(columns={'count_number': 'count_msno_total'}, inplace=True)
df4=df2.merge(df3,on='msno',how='left')
df4
```

```python
df4['total_avg']=df4['count_number']/df4['count_msno_total']
df4['w']=df4['count_number'] * df4['total_avg']
df4
```

```python
df5=df4.groupby('artist_id')['w'].sum()
df5=df5.reset_index()
df5.rename(columns={'w': 'sum_score'}, inplace=True)
df5
```

```python
df6=df4.merge(df5,on='artist_id',how='left')
df6=df6.drop(['count_number','count_msno_total','total_avg','w'],axis=1)
df6
```

```python
df7=df6.groupby('artist_id')['msno'].count()
df7=df7.reset_index()
df7.rename(columns={'msno': 'count_msno_number'}, inplace=True)
df7
```

```python
df6=df6.merge(df7,on='artist_id',how='left')
df6
```

```python
df6['score_avg']=df6['sum_score']/df6['count_msno_number']
df6
```

```python
df6.sort_values('score_avg',ascending=False)
```
