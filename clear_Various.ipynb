```python
import pandas as pd
import numpy as np
import math
from scipy.stats import entropy
import time
```

```python
df=pd.read_csv('finish_msno_rate',sep=',')
df
```

```python
df=df.drop(['total','rate'],axis=1)
df
```

```python
df=df[df['artist_id']!=484761]
df
```

```python
df['sum_count_number']=df.groupby("msno")["count_number"].transform("sum")
df
```

```python
df['count_number_rate']=df['count_number']/df['sum_count_number']
df['max_rate']=df.groupby("msno")["count_number_rate"].transform("max")
df
```

```python
df['each_artist_entrop']=df['max_rate'].apply(lambda n:(-n * np.log2(n))-((1-n)*np.log2(1-n)))
df
```

```python
df=df.fillna(0)
df
```
