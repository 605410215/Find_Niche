```python
import pandas as pd
import numpy as np
import math
from scipy.stats import entropy
import time
```

```python
df=pd.read_csv('msno_entropy_max_rate',sep=',')
df['each_artist_entrop']=df['each_artist_entrop'].round(2)
df=df[df['each_artist_entrop']<=0.1]
df
```

```python
df2=df.groupby(['artist_id','msno'])['count_number'].sum()
df2=df2.reset_index()
df2=df2.sort_values('artist_id',ascending=False)
df2
```

```python
df2['count_msno_total']=df2.groupby("msno")["count_number"].transform("sum")
df2
```

```python
df2['msno_proportion']=df2['count_number']/df2['count_msno_total']
df2['w']=df2['count_number'] * df2['msno_proportion']
df2
```

```python
df2['sum_score']=df2.groupby("artist_id")["w"].transform("sum")
#df2['sum_score2']=df2['sum_score'] * df2['msno_proportion']
df2
```

```python
df2['count_msno_number']=df2.groupby("artist_id")["msno"].transform("count")
df2
```

```python
df2['score_avg']=df2['sum_score']/df2['count_msno_number']
df2
```

```python
df2.sort_values('score_avg',ascending=False).drop_duplicates(['artist_id'], keep='last')
```
